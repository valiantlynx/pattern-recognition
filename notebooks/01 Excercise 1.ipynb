```python
# Statistical overview
desc = prices_and_sales_df.describe()

# Calculate the new statistic (e.g., Coefficient of Variation in percent)
# CV = (std / mean) * 100
cv_percent = (prices_and_sales_df.std() / prices_and_sales_df.mean()) * 100
variance_percent = (prices_and_sales_df.var() / (prices_and_sales_df.mean() ** 2)) * 100

# Add a new row named 'CV (%)' to the 'desc' DataFrame using .loc[]
desc.loc['Coefficient of Variation (%)'] = cv_percent
desc.loc['Variance (%)'] = variance_percent

print(desc)
```
