```python
# Statistical overview - select only numeric columns. not date etc.
numeric_df = prices_and_sales_df.select_dtypes(include='number')
desc = numeric_df.describe()

# CV = (std / mean) * 100
cv_percent = (numeric_df.std() / numeric_df.mean()) * 100
variance_percent = (numeric_df.var() / (numeric_df.mean() ** 2)) * 100

# Add a new row named 'CV (%)' to the 'desc' DataFrame using .loc[]
desc.loc['Coefficient of Variation (%)'] = cv_percent
desc.loc['Variance (%)'] = variance_percent

desc
```

on the data where the variance is high. sicne the variation is very high, it could mean that those as feature are stronger indicators than for example average price. we will see in the corelation matrix

```python
import plotly.express as px
import plotly.graph_objects as go

fig = px.imshow(numeric_df.corr(),
                labels=dict(x="Features", y="Features", color="Correlation"),
                x=numeric_df.corr().columns,
                y=numeric_df.corr().columns,
                color_continuous_scale='RdBu',
                zmin=-1,
                zmax=1)
fig
```

```python
# univariate analysis of 'Price' column
from scipy import stats
import numpy as np

fig = px.histogram(prices_and_sales_df, x='AveragePrice', 
                   nbins=prices_and_sales_df['AveragePrice'].nunique()//10,
                   title='Distribution of Price',
                   labels={'AveragePrice': 'Price'},
                   marginal='box',
                   histnorm='probability density')

# Calculate KDE for smooth density curve
data = prices_and_sales_df['AveragePrice'].dropna()
kde = stats.gaussian_kde(data)
x_range = np.linspace(data.min(), data.max(), 200)
y_kde = kde(x_range)

fig.add_trace(go.Scatter(x=x_range, y=y_kde, 
                         mode='lines', 
                         name='Density Curve',
                         line=dict(color='red', width=2)))

fig
```

```python
# univariate analysis of 'total volume' column
from scipy import stats
import numpy as np

# I wanna show the density but the deviation is too big. so i use the count to get the density
data = prices_and_sales_df['TotalVolume'].dropna()
log_data = np.log10(data)

fig = px.histogram(x=log_data, 
                   nbins=50,
                   title='Distribution of Total Volume (Log Scale)',
                   labels={'x': 'Log10(Total Volume)'},
                   marginal='box',
                   histnorm='probability density')

# Calculate KDE
kde = stats.gaussian_kde(log_data)
x_range = np.linspace(log_data.min(), log_data.max(), 200)
y_kde = kde(x_range)

fig.add_trace(go.Scatter(x=x_range, y=y_kde, 
                         mode='lines', 
                         name='Density Curve',
                         line=dict(color='red', width=2)))

fig.update_yaxes(title_text='Density')
fig
```
